# Завдання 2

## [Дані](https://www.kaggle.com/datasets/antongalysh/itstep-exam2/data)

Дані містять супутникові знімки, на частині яких знаходяться кораблі. Задача полягає у побудові моделі сегментації, яка зможе виявити корабель на зображенні. Результатом моделі(надалі маска зображення) має бути матриця такого ж розміру як і зображення з числами 0, 1 де 1 - піксель корабля, 0 - піксель не корабля

**Формат даних**
Усі зображення містяться в дерикторії `ship_images`, розмір зображення - (768, 768)

Також вам доступний датафрейм з двома стовчиками -- назвою файлу з зображенням та маскою зображення для одного корабля у форматі Run Length Encoding

Якщо на зображені немає корабля, то всі пікселі маски зображення рівні 0, а у відповідному стовпчику пропущене значення

Якщо на зображені декілька кораблів, то відповідна назва файлу у датафреймі зістрінеться декілька разів, і відповідна йому маска зображення буде різною, для кожного корабля окремо

**Run Length Encoding**
Для економії пам'яті результат моделі мають спеціальне кодування: кожні 2 числа позначають початкову позицію(індексація з 1) та кількість одиниць. Усі числа розділені пропуском.

Приклади:
* 2 5 означає що пікселі з індексами 2, 3, 4, 5, 6 рівні 1, а всі інші рівні 0
* 5 2 9 4 означає що пікселі з індексами 5, 6, 9, 10, 11, 12 рівні 1, а всі інші рівні 0

Індексація пікселів відбувається зверху вниз, а потім зліва вправо, тобто згідно індексації матриць 1 це (1,1), 2 - (2, 1), 3 - (3, 1), і так далі
